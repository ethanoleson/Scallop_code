```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as scipy
```

#### Functions


```python
def sphericity(length, width):
    sphere = length/width
    return sphere

#def sauter_mean(length):
    
```

```python
boy_scout = pd.read_csv('BSS_measurements.csv')
parks_ranch = pd.read_csv('PRC_measurements.csv')
boy_scout
```

```python
plt.hist(boy_scout['L(Downstream)'], bins=10, color='black')
plt.ylabel('Frequency')
plt.xlabel('Scallop Length (cm)')
plt.title('Boy Scout Springs Scallop Lengths')
```

```python
plt.hist(parks_ranch['L(Downstream)'], bins = 10)
plt.ylabel('Frequency')
plt.xlabel('Scallop Length (cm)')
plt.title('Parks Ranch Cave Scallop Lengths')
```

```python
boy_scout['sphericity'] = sphericity(boy_scout['L(Downstream)'], boy_scout['w(perp)'])
```

```python
parks_ranch['sphericity'] = sphericity(parks_ranch['L(Downstream)'], parks_ranch['w(perp)'])
```

```python
plt.hist(boy_scout['sphericity'], bins=10)
plt.ylabel('Frequency')
plt.xlabel('Sphericity')
plt.title('Boy Scout Springs Sphericity')
```

```python
plt.hist(parks_ranch['sphericity'], bins=10)
plt.ylabel('Frequency')
plt.xlabel('Sphericity')
plt.title('Parks Ranch Cave Sphericity')
```
