```python
import numpy as numpy
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

cm_data = pd.read_csv('cm_data_for_plots.csv')
cm_data


sns.scatterplot(data= cm_data, x= cm_data['feature'], y = cm_data['sa'], hue= cm_data['cave'],
                style= cm_data['cave'], markers={'Parks Ranch Cave': 'o', 'J2': 's', 'Schonberg-Hohlensystem': '^', 'Omega': '*'},
                palette='binary', s=100, edgecolor='black' )

plt.margins(x=0.5)
plt.xlabel('')
plt.ylabel('Sa $(\mu m)$')
plt.legend(title = '')

#plt.savefig('Sa_v_feature_by_cave.png', dpi= 300)
#plt.show()
```

```python
def roughnessplot(df, x_axis, color_by, marker_style, color_palette, markers_choice, save = None, rotation = None):
    df = spreadsheet
    sns.scatterplot(data= spreadsheet, x= spreadsheet[x_axis], y = spreadsheet['sa'], hue= spreadsheet[color_by],
                    style= spreadsheet[marker_style], markers=markers_choice,
                    palette=color_palette, s=100, edgecolor='black')
    plt.margins(x=0.4)
    plt.xticks(rotation = rotation)
    plt.xlabel('')
    plt.ylabel('Sa $(\mu m)$')
    plt.legend(title = '')
    plt.tight_layout()
    plt.savefig(save, dpi = 300)
    plt.show()
```
