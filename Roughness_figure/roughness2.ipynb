```python
import numpy as numpy
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

cm_data = pd.read_csv('cm_data_for_plots.csv')
cm_data


sns.scatterplot(data= cm_data, x= cm_data['feature'], y = cm_data['sa'], hue= cm_data['cave'],
                style= cm_data['cave'], markers={'Parks Ranch Cave': 'o', 'J2': 's', 'Schonberg-Hohlensystem': '^', 'Omega': '*'},
                palette='binary', s=100, edgecolor='black' )
plt.margins(x=0.5)
plt.xlabel('')
plt.ylabel('Sa $(\mu m)$')
plt.legend(title = '')

#plt.savefig('Sa_v_feature_by_cave.png', dpi= 300)
#plt.show()
```

```python
sns.scatterplot(data= cm_data, x= cm_data['cave'], y = cm_data['sa'], hue= cm_data['feature'],
                style= cm_data['feature'], markers={'Mechanical': 'o', 'Solutional': 's'},
                palette='binary', s=100, edgecolor='black' )
plt.margins(x=0.4)
plt.xticks(rotation=45)
plt.xlabel('')
plt.ylabel('Sa $(\mu m)$')
plt.legend(title = '')
plt.tight_layout()
plt.savefig('Sa_v_cave_by_feature.png', dpi = 300)
plt.show()
```
