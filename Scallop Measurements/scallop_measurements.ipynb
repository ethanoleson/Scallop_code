```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as scipy
import seaborn as sns
```

#### Functions


```python
def aspect_ratio(length, width):
    aspect = length/width
    return aspect

def sauter_mean(length):
    s_mean = ((sum(length**3))/(sum(length**2)))
    return s_mean
    
```

```python
boy_scout = pd.read_csv('BSS_measurements.csv')
parks_ranch = pd.read_csv('PRC_measurements.csv')
parks_ranch['L(Downstream)']
```

```python
## Arithmetic mean:

bss_a_mean = boy_scout['L(Downstream)'].mean()
prc_a_mean = parks_ranch['L(Downstream)'].mean()

## Sauter Mean: 

bss_s_mean = sauter_mean(boy_scout['L(Downstream)'])
prc_s_mean = sauter_mean(parks_ranch['L(Downstream)'])
```

```python

sns.histplot(boy_scout, x = boy_scout['L(Downstream)'], bins=10, color= 'black', alpha = 0.5)
plt.plot([bss_a_mean, bss_a_mean], [0, 7], linewidth = 2, color = 'red', label = 'Arithmetic Mean')
plt.plot([bss_s_mean, bss_s_mean], [0, 7], linewidth = 2, color = 'blue', label = 'Sauter Mean')
plt.text(8, 5.5, 'n=38', fontsize = 14)
plt.ylabel('Frequency', fontsize = 14)
plt.xlabel('Scallop Length (cm)', fontsize = 14)
plt.legend()
#plt.savefig('boy_scout_scallop_lengths.png', dpi = 600)
plt.show()
```

```python
sns.histplot(parks_ranch, x = parks_ranch['L(Downstream)'], bins=10, color= 'gray', alpha = 0.5)
plt.plot([prc_a_mean, prc_a_mean], [0, 8], linewidth = 2, color = 'red', label = 'Arithmetic Mean')
plt.plot([prc_s_mean, prc_s_mean], [0, 8], linewidth = 2, color = 'blue', label = 'Sauter Mean')
plt.text(1.5, 6.5, 'n=39', fontsize = 14)
plt.ylabel('Frequency', fontsize = 14)
plt.xlabel('Scallop Length (cm)', fontsize = 14)
plt.legend()
#plt.savefig('prc_scallop_lengths.png', dpi = 600)
plt.show()
```

```python
boy_scout['aspect_ratio'] = aspect_ratio(boy_scout['L(Downstream)'], boy_scout['w(perp)'])
```

```python
parks_ranch['aspect_ratio'] = aspect_ratio(parks_ranch['L(Downstream)'], parks_ranch['w(perp)'])
```

```python
sns.histplot(boy_scout, x= boy_scout['aspect_ratio'], bins = 10,  color= 'black', alpha = 0.5)
#plt.ylabel('Frequency',fontsize = 14)
plt.ylabel('')
plt.xlabel('Aspect Ratio', fontsize = 14)
#plt.savefig('bss_aspect_ratio', dpi = 600)
plt.show()
```

```python
sns.histplot(parks_ranch, x= parks_ranch['aspect_ratio'], bins = 10, color= 'gray', alpha = 0.5)
#plt.ylabel('Frequency', fontsize = 14)
plt.ylabel('')
plt.xlabel('Aspect Ratio',fontsize = 14)
#plt.savefig('prc_aspect_ratio', dpi = 600)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'data' with columns 'feature1', 'feature2', 'feature3', etc.
# Replace these with your actual column names and dataset.

# Create a sample DataFrame for demonstration purposes
import pandas as pd
import numpy as np

np.random.seed(42)
data = pd.DataFrame({
    'feature1': np.random.randn(1000),
    'feature2': np.random.randn(1000),
    'feature3': np.random.randn(1000),
    'feature4': np.random.randn(1000),
    'feature5': np.random.randn(1000),
    'feature6': np.random.randn(1000),
})

# Set up the subplots
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(12, 8))

# Flatten the 2D array of subplots to simplify indexing
axes = axes.flatten()

# Loop through each subplot and plot the histogram
for i, column in enumerate(data.columns):
    sns.histplot(data[column], ax=axes[i], kde=True, color='skyblue')
 

    # Add x-labels only to the subplots in the third row
    if i >= 3:
        axes[i].set_xlabel('Value')
    else:
        axes[i].set_xlabel('')  # Set blank x-label for other subplots

    # Add y-labels only to the subplots in the first column
    if i % 3 == 0:
        axes[i].set_ylabel('Frequency')
    else:
        axes[i].set_ylabel('')  # Set blank y-label for other subplots

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```
