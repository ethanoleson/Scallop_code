```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as scipy
```

```python
grain_size = pd.read_csv('Oleson_grain_size_mesure_master.csv')
grain_size
```

```python
Omega71923B = grain_size[grain_size['Sample'] == '71923B']
Omega71923A = grain_size[grain_size['Sample']== '71923A']
BMTC80823A  = grain_size[grain_size['Sample'] == '80823A']
```

```python
BMTC80823A
```

### Omega Limestone - 71923A


```python
Omega71923A_mean = Omega71923A['height'].mean()
Omega71923B_mean = Omega71923B['height'].mean()
BMTC80823A_mean = BMTC80823A['height'].mean()

Omega71923A_mean
```

```python
fig, ax = plt.subplots()
sns.histplot(x= Omega71923A['height'], color= 'lightgray')
plt.xlabel('Grain Size ($\mu m$)')
plt.plot([Omega71923A_mean, Omega71923A_mean], [0, 23], color = 'black', linewidth = 2.0)

#plt.savefig('OmegaLS_grain_size.png', dpi = 300)
#plt.show()
```
