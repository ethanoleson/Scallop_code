```python
Omega71923B = grain_size[grain_size['Sample'] == '71923B']
Omega71923A = grain_size[grain_size['Sample']== '71923A']
BMTC80823A  = grain_size[grain_size['Sample'] == '80823A']
prc = grain_size[grain_size['Sample'] == 'prc']
j2 = grain_size[grain_size['Sample'] == 'j2']
lpns = grain_size[grain_size['Sample']== 'lpns']
lpsc = grain_size[grain_size['Sample'] == 'lpsc']
onon83123A = grain_size[grain_size['Sample']== '83123A']
Bss90423A = grain_size[grain_size['Sample']== '90423A']


Omega71923A_mean = Omega71923A['Height (µm)'].mean()
Omega71923B_mean = Omega71923B['Height (µm)'].mean()
BMTC80823A_mean = BMTC80823A['Height (µm)'].mean()
prc_mean = prc['Height (µm)'].mean()
j2_mean = j2['Height (µm)'].mean()
lpns_mean = lpns['Height (µm)'].mean()
lpsc_mean = lpsc['Height (µm)'].mean()
onon83123A_mean =  onon83123A['Height (µm)'].mean()
Bss90423A_mean = Bss90423A['Height (µm)'].mean()


Omega71923A_max = Omega71923A['Height (µm)'].max()
Omega71923B_max = Omega71923B['Height (µm)'].max()
BMTC80823A_max = BMTC80823A['Height (µm)'].max()
prc_max = prc['Height (µm)'].max()
j2_max = j2['Height (µm)'].max()
lpns_max = lpns['Height (µm)'].max()
lpsc_max = lpsc['Height (µm)'].max()
onon83123A_max =  onon83123A['Height (µm)'].max()
Bss90423A_max = Bss90423A['Height (µm)'].max()


Omega71923A_min = Omega71923A['Height (µm)'].min()
Omega71923B_min = Omega71923B['Height (µm)'].min()
BMTC80823A_min = BMTC80823A['Height (µm)'].min()
prc_min = prc['Height (µm)'].min()
j2_min= j2['Height (µm)'].min()
lpns_min = lpns['Height (µm)'].min()
lpsc_min = lpsc['Height (µm)'].min()
onon83123A_min =  onon83123A['Height (µm)'].min()
Bss90423A_min = Bss90423A['Height (µm)'].min()

Omega71923A_median = Omega71923A['Height (µm)'].median()
Omega71923B_median = Omega71923B['Height (µm)'].median()
BMTC80823A_median = BMTC80823A['Height (µm)'].median()
prc_median = prc['Height (µm)'].median()
j2_median = j2['Height (µm)'].median()
lpns_median = lpns['Height (µm)'].median()
lpsc_median = lpsc['Height (µm)'].median()
onon83123A_median = onon83123A['Height (µm)'].median()
Bss90423A_median = Bss90423A['Height (µm)'].median() 

print('Omega 71921A', "mean, median,max,min:", Omega71923A_mean, Omega71923A_median, Omega71923A_max, Omega71923A_min)
print('Omega 71921B', "mean, median, max,min:", Omega71923B_mean, Omega71923B_median, Omega71923B_max, Omega71923B_min)
print('BM-TC', "mean,median, max,min:", BMTC80823A_mean, BMTC80823A_median, BMTC80823A_max, BMTC80823A_min)
print('prc', "mean, median, max,min:", prc_mean, prc_median, prc_max, prc_min)
print('j2', "mean,median, max,min:", j2_mean, j2_median, j2_max, j2_min)
print('lpns', "mean, median, max,min:", lpns_mean, lpns_median, lpns_max, lpns_min)
print('lpsc', 'mean, median, max, min', lpsc_mean, lpsc_median, lpsc_max, lpsc_min)
print('Onondaga', "mean, median, max,min:", onon83123A_mean, onon83123A_median, onon83123A_max, onon83123A_min)
print('BSS', "mean, median, max,min:", Bss90423A_mean, Bss90423A_median, Bss90423A_max, Bss90423A_min)
```

```python
violin_data = pd.read_csv('Oleson_grain_size_forViolin.csv')
violin_data['Sample'].unique()
```

```python
violinplot_pallete = {"Scalloped": "lightgray", "Non-scalloped": "dimgray"}
```

```python
fig, ax = plt.subplots()
sns.violinplot(violin_data, y = 'Height', x = 'Sample', ax= ax, hue= 'scalloped',palette= violinplot_pallete, dodge=False, order=['71923A', '71923B', 'LPSC', 'LPNS', 'J2', '80823A', '111123A', '83123A', '90423A' ])
plt.xticks(rotation = 45)
ax.set_xlabel('')
ax.set_ylabel('Intermediate Length ($\mu$m)')
plt.legend(title = '')
ax.set_yscale('log')
plt.tight_layout()
plt.savefig('grain_size_violin_plots.png', dpi =600, bbox_inches = 'tight' )
plt.show()
```
