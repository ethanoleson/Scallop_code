```python
import matplotlib.pyplot as plt

def plot_on_number_line(points):
    plt.axhline(y=0, color='black', linestyle='-')  # Horizontal line representing the number line
    
    # Plotting points on the number line
    for point in points:
        plt.scatter(point, 0, color='red', marker='o')  # Plotting points in red color
    
    plt.title('Number Line')
    plt.xlabel('Point')
    plt.grid(True)
    plt.show()

# Example usage
points_to_plot = [1, 3, -2, 5]
plot_on_number_line(points_to_plot)
```

```python
def modal_percentage(df, counts):
    sample = df.sample    
    calcite = df.calcite
    dolomite = df.dolomite
    clay = df.clay
    gypsum = df.gypsum
    sio2 = df.silica
    other = df.other
    percent_calcite = calcite/counts * 100
    percent_dolo = dolomite/counts * 100 
    percent_gypsum = gypsum/counts * 100
    percent_clay = clay/counts *100
    percent_sio2 = sio2/counts *100
    percent_other = other/counts *100

    return percent_calcite, percent_dolo, percent_gypsum, percent_clay, percent_sio2, percent_other

```

```python
import matplotlib.pyplot as plt

# Create a number line from 0 to 100
number_line_values = list(range(101))

# Set the labels at intervals of 0, 10, 50, 90, and 100
labels = [0, 10, 50, 90, 100]

# Create the plot
plt.plot(number_line_values, [0] * len(number_line_values), color='black', linewidth=2, marker='|', markersize=10)

# Add labels at specified intervals with adjusted vertical position
for label in labels:
    plt.text(label, 0.1, str(label), fontsize=12, ha='center', va='center')

# Customize the plot
plt.xlim(0, 100)
plt.yticks([])  # Remove y-axis ticks
plt.title('Number Line from 0 to 100')
plt.xlabel('Value')

# Show the plot
plt.show()
```
