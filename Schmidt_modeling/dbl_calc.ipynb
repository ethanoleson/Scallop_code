```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
calcium_charge = 2
carbonate_charge = 2
Ca_diff = 0.00000793 #cm^2 s^-1
CaCO3_diff = 0.00000955 #cm^2 s^-1

def diff_co(z_cat, z_an, D_cat, D_an):
    num = (z_cat + z_an)*D_cat*D_an
    denom = z_cat*D_cat + z_an*D_an
    diff = num/denom
    return diff
```

```python
Calcite_diff = diff_co(calcium_charge, carbonate_charge, Ca_diff, CaCO3_diff)
Calcite_diff
```

```python
D_h = np.linspace(0.001, 1, 1000)
mu = 1.00 #cm^-1*g*s^-1
rho = 1.00 #g*cm^-3
f= 0.005
Re = 2200
D = Calcite_diff

def schmidt(mu, rho, D):
    schmidt = mu/(rho*D)
    return schmidt

def sherwood(f,Re,Sc):
    a = 8
    b = 1000
    c = 1 + 12.7*(np.sqrt(f/a))
    d = Sc**(2/3) - 1

    sherwood = ((f/a)*(Re - b)*Sc)/(c*d)

    return sherwood

def dbl(D_h,Sh):
    dbl = D_h/Sh
    return dbl



```

```python
calcite_sc = schmidt(mu, rho, Calcite_diff)
calcite_sh = sherwood(f, Re, calcite_sc)
calcite_dbl  = dbl(1, calcite_sh )
calcite_dbl
```
